```python
import pandas as pd

df = pd.read_csv(
    "co2_raw_data.csv",
    sep=",",
    encoding="utf-8",
    low_memory=False
)

print(df.head())
```

```python
columns_needed = [
    "country",
    "year",
    "population",
    "gdp",
    "co2",
    "co2_per_capita",
    "coal_co2",
    "oil_co2",
    "gas_co2",
    "cement_co2"
]

df_clean = df[columns_needed]

df_clean = df_clean[df_clean["year"] >= 1990]

df_clean.to_csv("co2_cleaned.csv", index=False)
df_clean
```
